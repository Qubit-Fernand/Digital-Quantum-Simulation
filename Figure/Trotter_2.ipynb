```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# 设置全局字体为 Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

# 启用 LaTeX 渲染
plt.rcParams['text.usetex'] = True

r_50 = loadmat('r_50.mat')['r_50'].flatten()
r_50_Delta_5 = loadmat('r_50_Delta_5.mat')['r_50_Delta_5'].flatten()
r_100 = loadmat('r_100.mat')['r_100'].flatten()
r_100_Delta_5 = loadmat('r_100_Delta_5.mat')['r_100_Delta_5'].flatten()
r_200 = loadmat('r_200.mat')['r_200'].flatten()
r_200_Delta_5 = loadmat('r_200_Delta_5.mat')['r_200_Delta_5'].flatten()
r_500 = loadmat('r_500.mat')['r_500'].flatten()
r_500_Delta_5 = loadmat('r_500_Delta_5.mat')['r_500_Delta_5'].flatten()
```

```python
index = [0.005, 0.008, 0.010, 0.012, 0.015, 0.017, 0.020]
index_new = np.linspace(min(index), max(index), 100)

# 多项式拟合
coefficients_r_50 = np.polyfit(index, r_50, 3) 
poly_r_50 = np.poly1d(coefficients_r_50)
coefficients_r_5_Delta_5 = np.polyfit(index, r_50_Delta_5, 3) 
poly_r_50_Delta_5 = np.poly1d(coefficients_r_5_Delta_5)

coefficients_r_100 = np.polyfit(index, r_100, 3)  
poly_r_100 = np.poly1d(coefficients_r_100)
coefficients_r_100_Delta_5 = np.polyfit(index, r_100_Delta_5, 3)  
poly_r_100_Delta_5 = np.poly1d(coefficients_r_100_Delta_5)

coefficients_r_200 = np.polyfit(index, r_200, 3) 
poly_r_200 = np.poly1d(coefficients_r_200)
coefficients_r_200_Delta_5 = np.polyfit(index, r_200_Delta_5, 3) 
poly_r_200_Delta_5 = np.poly1d(coefficients_r_200_Delta_5)

coefficients_r_500 = np.polyfit(index, r_500, 3) 
poly_r_500 = np.poly1d(coefficients_r_500)
coefficients_r_500_Delta_5 = np.polyfit(index, r_500_Delta_5, 3)  
poly_r_500_Delta_5 = np.poly1d(coefficients_r_500_Delta_5)
```

```python
fig, ax = plt.subplots(dpi=270)
ax.set_title('Trotter Error in the Low Energy Subspace $\Delta = 5$')
ax.set_xlabel('Single Step Evolution Time $s$')
ax.set_ylabel('Trotter Error $\epsilon$')

ax.scatter(index, r_50, marker = '*', color = '#1f77b4', label='$r=50$')
ax.plot(index_new, poly_r_50(index_new), color = '#1f77b4', linestyle = '-')

ax.scatter(index, r_50_Delta_5, marker ='*', color = '#1f77b4')
ax.plot(index_new, poly_r_50_Delta_5(index_new), color = '#1f77b4', linestyle = '--')

ax.scatter(index, r_100, marker = '*', color = '#ff7f0e', label='$r=100$')
ax.plot(index_new, poly_r_100(index_new), color = '#ff7f0e', linestyle = '-')

ax.scatter(index, r_100_Delta_5, marker ='*', color = '#ff7f0e')
ax.plot(index_new, poly_r_100_Delta_5(index_new), color = '#ff7f0e', linestyle = '--')

ax.scatter(index, r_200, marker = '*', color = '#2ca02c', label='$r=200$')
ax.plot(index_new, poly_r_200(index_new), color = '#2ca02c', linestyle = '-')

ax.scatter(index, r_200_Delta_5, marker ='*', color = '#2ca02c')
ax.plot(index_new, poly_r_200_Delta_5(index_new), color = '#2ca02c', linestyle = '--')

ax.scatter(index, r_500, marker = '*', color = '#d62728', label='$r=500$')
ax.plot(index_new, poly_r_500(index_new), color = '#d62728', linestyle = '-')

ax.scatter(index, r_500_Delta_5, marker ='*', color = '#d62728')
ax.plot(index_new, poly_r_500_Delta_5(index_new), color = '#d62728', linestyle = '--')
ax.legend()
```
